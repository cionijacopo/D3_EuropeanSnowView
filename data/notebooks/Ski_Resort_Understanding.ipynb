```python
import numpy as np
import pandas as pd 
import os
import pycountry
```

```python
ski_resort = pd.read_csv("../European_Ski_Resorts.csv", low_memory=False)
```

```python
ski_resort.info()
ski_resort.head(10)
```

```python
ski_duplicates = ski_resort.duplicated().sum()
print(f"Numero di duplicati nel dataframe: {ski_duplicates}")
```

```python
ski_resort.describe()
```

```python
# Ripulisco dalla prima colonna
ski_resort = ski_resort.drop(columns=["Unnamed: 0"])
ski_resort.head(10)
```

```python
# Conteggio dei valori unici nella colonna 'Country'
unique_countries = ski_resort['Country'].nunique()
print(f"Paesi unici nel dataframe: {unique_countries}")
```

```python
# Conteggio dei record per ogni paese 
cont = ski_resort['Country'].value_counts()
cont
```

```python
# Crea una cartella temporanea per salvare i file CSV per ogni paese
output_folder = "./resorts_by_country"
os.makedirs(output_folder, exist_ok=True)

# Per ogni paese, salva un file CSV con i suoi record
for country, group in ski_resort.groupby("Country"):
    # Sostituisce eventuali caratteri problematici nel nome file
    safe_country = country.replace(" ", "_").replace("/", "_")
    file_path = os.path.join(output_folder, f"{safe_country}.csv")
    group.to_csv(file_path, index=False)
```

```python
# Conta quanti resort ci sono per ciascun paese (per nome)
resort_counts = ski_resort['Country'].value_counts().reset_index()
resort_counts.columns = ['country_name', 'resorts_count']

#Funzione per ottenere il codice ISO A2 (IT, FR, etc.)
def get_country_code(name):
    try:
        return pycountry.countries.lookup(name).alpha_2
    except LookupError:
        return None

#Applica la funzione
resort_counts['country_code'] = resort_counts['country_name'].apply(get_country_code)

#Rimuovi righe con country_code mancante
resort_counts = resort_counts.dropna(subset=['country_code'])

# Rimuovo la colonna country_name
resort_counts.drop(columns=['country_name'], inplace=True)

# Aggiungo manualmente la russia
new_row = {
    'country_code': 'RU',     # codice del nuovo paese
    'resorts_count': 4        # numero di resort nel paese
}

resort_counts.loc[len(resort_counts)] = new_row


#Salva il file finale
resort_counts[['country_code', 'resorts_count']].to_csv("resorts.csv", index=False)

#Visualizza i primi risultati
resort_counts.head()
```
